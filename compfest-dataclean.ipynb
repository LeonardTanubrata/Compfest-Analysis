## Create a 1x2 grid of subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=df, x=df['likes_count'], ax=axes[0])
axes[0].set_title('Likes Count')
sns.boxplot(data=df, x=df['comments_count'], ax=axes[1])
axes[1].set_title('Comments Count')
```

Seeing from the plot above that the two columns are not normally distributed, we will do a simple filling using the median while for the content type we fill it with the mode because the column is categorical.

```python
median = df['likes_count'].median()
df['likes_count'].fillna(median,inplace=True)
# %%
com_median = df['comments_count'].median()
df['comments_count'].fillna(com_median,inplace=True)
# %%
mode = df['content_type'].mode()[0]
df['content_type'].fillna(mode,inplace=True)
```

## Case 2

### 1. Slicing Dataframe with post_time column before 31 july 2023

```python
df_slice = df[df['post_time'] <= '2023-07-31'].sort_values('post_time')
df_slice
```

### 2. Searching for anyone whose caption has the specified hashtags

```python
# Define the list of hashtags to check for
hashtags_to_check = ['#Academy', '#Competition', '#EvolveForTomorrow', '#CF15', '#COMPFEST15']

# Use str.contains() to filter rows containing any of the specified hashtags
filtered_df = df_slice[df_slice['caption'].str.contains('|'.join(hashtags_to_check))]

# Display the filtered DataFrame
filtered_df
```

### 3. Find whose username had tagged 3 or more of their friends

```python
# Define regular expressions for matching "@" and "#" with content
at_with_content_regex = r'@([\w\d]+)'

# Extract "@" mentions with content
filtered_df['mentions'] = filtered_df['caption'].str.extractall(at_with_content_regex).groupby(level=0)[0].apply(', '.join)

filtered_df.dropna(inplace=True)

# List of words to remove
words_to_remove = ['compfest,', 'compest,', 'compfst,','compefst,','COMPFEST','compfest','compest','compfst','compefst']

# Function to remove multiple words
def remove_words(text, words):
    for word in words:
        text = text.replace(word, '')
    return text

# Use the remove_words function to remove multiple words
filtered_df['friend mention'] = filtered_df['mentions'].apply(lambda x: remove_words(x, words_to_remove))

# Count the number of mentions in each row
filtered_df['mention_count'] = filtered_df['friend mention'].str.count(',') + 1

# Filter rows with more than 4 mentions
filtered_df = filtered_df[filtered_df['mention_count'] >= 3 ]

filtered_df.head()
```

### 4. Make new column for engagement rate

```python
filtered_df['engagement'] = filtered_df['likes_count'] + filtered_df['comments_count']
filtered_df.sort_values('engagement',ascending=False).head(30).reset_index().drop(columns='index').head()
```
